```python
class Student:

    def __init__(self, name: str, student_id: str):
        self.name = name
        self.student_id = student_id
        self.score_list = []

    def add_score(self, *score):
        try:
            for i in score:
                if 0 <= i <= 100:
                    self.score_list.append(i)
                    print(f"Score {i} added for {self.name}")
                else:
                    print("Score must be between 0 and 100")
            return "All valid scores have been added"
        except ValueError:
            print("Error: Invalid Score")

        

    def calculate_average(self):
        average = sum(self.score_list) / len(self.score_list)
        return average

    def get_grade(self):
        average_score = self.calculate_average()

        if average_score >= 70:
            return "Grade A"
        elif average_score >= 60:
            return "Grade B"
        elif average_score >= 40:
            return "Grade C"
        elif average_score >= 30:
            return "Grade D"
        else:
            return "Grade F"

    def print_report(self):
        print(f"Report for {self.name} of student Id of {self.student_id}")
        print(f"Scores : {self.score_list}")
        print(f"Average: {self.calculate_average()}")
        print(f"Grade : {self.get_grade()}")



jennifer = Student("jennifer", "ID001")
elo = Student("elo", "ID002")
ella = Student("ella", "ID003")

jennifer.add_score(85,60,70,75,90,100,90)

jennifer.calculate_average()
jennifer.print_report()

elo.add_score(70,50,50,77,50,90,74,)


elo.calculate_average()
elo.print_report()

ella.add_score(56,80,43,78,65,88,70)

ella.calculate_average()
ella.print_report() 

    
```
